# text cleaning

sample 10k rows
remove html tages,special chars,to lower case,stop words removal,stemminmg


```python
df=df.sample(10000)
```

```python
df.info()
```

```python
df['sentiment'].replace({'positive':1,"negative":0},inplace=True)
```

```python
df
```

```python
# remove html tags

import re
clean=re.compile('<.*?>')
re.sub(clean,'',df.iloc[2].review)
```

```python
# function to clean the html tages

def clean_html(text):
    clean=re.compile('<.*?>')
    return re.sub(clean,"",text)
```

```python
df['review']=df['review'].apply(clean_html)
```

```python
df
```

```python
# converting everythin to lower

def convert_lower(text):
    return text.lower()
```

```python
df['review']=df['review'].apply(convert_lower)
```

```python
# function to remove the special chars

def remove_special(text):
    x=''
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+" "
    return x
```

```python
df['review']=df['review'].apply(remove_special)
```

```python
# removing the stop words
import nltk
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
# function to remove the stops words
def remove_stopwords(text):
    x=[]
    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    y=x[:]
    x.clear()
    return y
```

```python
df['review']=df['review'].apply(remove_stopwords)
```

```python
df
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
y=[]
def stem_words(text):
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear()
    return z
```

```python
df['review']=df['review'].apply(stem_words)
```

```python
#join back
def join_back(list_input):
    return " ".join(list_input)
```

```python
df['review']=df['review'].apply(join_back)
```
